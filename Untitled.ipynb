```python
import matplotlib.pyplot as plt
import numpy as np

data = [
    ["13:51", 345],
    ["14:16", 338],
    ["15:11", 315],
    ["15:26", 310],
    ["15:33", 306],
    ["15:44", 301],
    ["15:52", 310],
    ["15:55", 313],
    ["15:58", 293],
    ["16:29", 287],
    ["16:34", 295],
    ["16:34", 271],
    ["16:35", 279],
    ["16:51", 278],
    ["16:53", 271],
    ["16:53", 272],
    ["17:07", 270],
    ["17:10", 261],
    ["17:17", 268],
    ["17:28", 266],
    ["17:34", 257],
    ["17:34", 251],
    ["17:39", 251],
    ["19:19", 224],
    ["19:44", 223]
]

pdata = [[], []]
for i in data:
    tm = int(i[0].split(":")[0]) * 60 + int(i[0].split(":")[1])
    pdata[0].append(i[1])
    pdata[1].append(tm)

plt.scatter(pdata[0], pdata[1])

a, b = np.polyfit(pdata[0], pdata[1], 1)

pts = [[], []]
for i in range(200, 400):
    pts[0].append(i)
    pts[1].append(a * i + b)

for i in range(600, 1201, 60):
    print(str(i//60) + ":00", "\t", int((i - b) // a))
    

plt.plot(pts[0], pts[1])
plt.show()

print(a, b)
```
